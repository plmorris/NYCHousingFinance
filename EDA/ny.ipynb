```python
import pandas as pd
import numpy as np
```

```python
linkQ ='https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_queens.xlsx'
queens = pd.read_excel(linkQ, skiprows=4)
linkB ='https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_brooklyn.xlsx'
bkl = pd.read_excel(linkB, skiprows=4)
linkM ='https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_manhattan.xlsx'
manhattan = pd.read_excel(linkM, skiprows=4)
linkS ='https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_statenisland.xlsx'
si = pd.read_excel(linkS, skiprows=4)
linkBr ='https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_bronx.xlsx'
bronx = pd.read_excel(linkBr, skiprows=4)
```

```python
queens = queens.drop(['EASEMENT', 'APARTMENT NUMBER','TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT'], axis=1)
bkl = bkl.drop(['EASEMENT', 'APARTMENT NUMBER','TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT'], axis=1)
manhattan = manhattan.drop(['EASEMENT', 'APARTMENT NUMBER','TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT'], axis=1)
si = si.drop(['EASEMENT', 'APARTMENT NUMBER','TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT'], axis=1)
bronx = bronx.drop(['EASEMENT', 'APARTMENT NUMBER','TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT'], axis=1)
```

```python
queens['GROSS SQUARE FEET'].replace(0.0, np.nan,inplace=True)
queens['LAND SQUARE FEET'].replace(0.0, np.nan,inplace=True)
queens['TOTAL UNITS'].replace(0.0, np.nan,inplace=True)
queens['SALE PRICE'].replace(0.0, np.nan,inplace=True)
queens['YEAR BUILT'].replace(0.0, np.nan,inplace=True)
bkl['GROSS SQUARE FEET'].replace(0.0, np.nan,inplace=True)
bkl['LAND SQUARE FEET'].replace(0.0, np.nan,inplace=True)
bkl['TOTAL UNITS'].replace(0.0, np.nan,inplace=True)
bkl['SALE PRICE'].replace(0.0, np.nan,inplace=True)
bkl['YEAR BUILT'].replace(0.0, np.nan,inplace=True)
manhattan['GROSS SQUARE FEET'].replace(0.0, np.nan,inplace=True)
manhattan['LAND SQUARE FEET'].replace(0.0, np.nan,inplace=True)
manhattan['TOTAL UNITS'].replace(0.0, np.nan,inplace=True)
manhattan['SALE PRICE'].replace(0.0, np.nan,inplace=True)
manhattan['YEAR BUILT'].replace(0.0, np.nan,inplace=True)
si['GROSS SQUARE FEET'].replace(0.0, np.nan,inplace=True)
si['LAND SQUARE FEET'].replace(0.0, np.nan,inplace=True)
si['TOTAL UNITS'].replace(0.0, np.nan,inplace=True)
si['SALE PRICE'].replace(0.0, np.nan,inplace=True)
si['YEAR BUILT'].replace(0.0, np.nan,inplace=True)
bronx['GROSS SQUARE FEET'].replace(0.0, np.nan,inplace=True)
bronx['LAND SQUARE FEET'].replace(0.0, np.nan,inplace=True)
bronx['TOTAL UNITS'].replace(0.0, np.nan,inplace=True)
bronx['SALE PRICE'].replace(0.0, np.nan,inplace=True)
bronx['YEAR BUILT'].replace(0.0, np.nan,inplace=True)
```

```python
queens = queens.dropna()
bkl = bkl.dropna()
manhattan = manhattan.dropna()
si = si.dropna()
bronx = bronx.dropna()
```

```python
ny = pd.concat([queens, bkl, manhattan, si, bronx], ignore_index=True)
```
